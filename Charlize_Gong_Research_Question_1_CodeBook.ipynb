Codebook 1

Variable: H4WP7Screenshot 2023-04-28 at 2.37.39 PM.png

```python
#Manage data for variable 1
df['H4WP7'].value_counts()
```

```python
df['father_alive'] = df['H4WP7']
df['father_alive'] = df['father_alive'].loc[df['father_alive'] != ' ']
df['father_alive'] = df['father_alive'].loc[df['father_alive'] != '6']
df['father_alive'] = df['father_alive'].loc[df['father_alive'] != '8']
df['father_alive'].value_counts()
```

```python
df['father_alive'].replace(['0','1'],['NO','YES'],inplace = True)
df['father_alive'].value_counts()
```

```python
rp.summary_cat(df['father_alive'])
```

```python
#Graphs and tables or statistics
```

```python
plt.title("Is your biological father still alive?")
explode = (0.1, 0.2)
df['father_alive'].value_counts().plot(kind= 'pie', autopct='%1.1f%%',  explode=explode, shadow=True, startangle=90)
```

```python
df['father_alive'].value_counts(normalize = True).plot(kind ='bar')
plt.title("Is your biological father still alive?")
plt.ylabel("Percent")
```

```python
#Manage data for variable 1
df['H4WS1'].dtype
```

```python
df['num_sib'] = df['H4WS1']
df['num_sib'].dtype
```

```python
df['num_sib'] = pd.to_numeric( df['num_sib'] , errors='coerce' )
df['num_sib'].dtype
```

```python
df['num_sib'] = df['num_sib'].loc[ df['num_sib'] != ' ' ]
df['num_sib'] = df['num_sib'].loc[ df['num_sib'] != '96' ]
df['num_sib'] = df['num_sib'].loc[ df['num_sib'] != '98' ]
df['num_sib'] = df['num_sib'].loc[ df['num_sib'] < 21 ]
df['num_sib'].value_counts()
```

```python
df['num_sib'].value_counts(normalize=True).round(4)
```

```python
freq_table = pd.crosstab(df['num_sib'], 'percentage', normalize=True).round(4)
freq_table
```

```python
df['num_sib'].describe()
```

```python
rp.summary_cont(df['num_sib'])
```
